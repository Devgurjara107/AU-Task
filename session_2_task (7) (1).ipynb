### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
sal=int(input("enter the salary :"))
if sal<=499999:
  print("after deduction of 0 percent tax =",sal)
elif sal>=500000 and sal<=999999:
  saltax=sal*(10/100)
  final_salary=sal-saltax
  print("after deduction of 10 percent tax =",final_salary)
elif sal>=1000000 and sal<=1999999:
  saltax=sal*(20/100)
  final_salary=sal-saltax
  print("after deduction of 20 percent tax =",final_salary)
else:
  saltax=sal*(30/100)
  final_salary=sal-saltax
  print("after deduction of 30 percent tax =",final_salary)
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Write code here
num=int(input("enter the number"))
facto=1
i=num
while i>0:
  facto=facto*i
  i-=1
print(facto)
````
